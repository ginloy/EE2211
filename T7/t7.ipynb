```python
from custom.models import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([-10, -8, -3, -1, 2, 7]).reshape(-1, 1)
y_train = np.array([4.18, 2.42, 0.22, 0.12, 0.25, 3.09]).reshape(-1, 1)
x_test = np.array([-9, -7, -5, -4, -2, 1, 4, 5, 6, 9]).reshape(-1, 1)
y_test = np.array([3, 1.81, 0.8, 0.25, -0.19, 0.4, 1.24, 1.68, 2.32, 5.05]).reshape(-1, 1)
```

```python
powers = np.arange(1, 7, 1).reshape(-1, 1)
def fit_and_test(power):
    model = PolyModel(x_train, y_train, power)
    y_pred = model.predict(x_test)
    mse = np.mean((y_pred - y_test) ** 2)
    return mse
mses = np.vectorize(fit_and_test)(powers)
plt.plot(powers, mses)
plt.show()
```

```python
powers = np.arange(1, 7, 1).reshape(-1, 1)
def fit_and_test(power):
    model = PolyModel(x_train, y_train, power, ridge=1)
    y_pred = model.predict(x_test)
    mse = np.mean((y_pred - y_test) ** 2)
    return mse
mses = np.vectorize(fit_and_test)(powers)
plt.ylim(0, 10)
plt.plot(powers, mses)
```

There are lower errors for especially for higher orders
