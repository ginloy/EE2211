# Qn 2

```python
df_omnibuses = pd.read_csv("annual-motor-vehicle-population-by-vehicle-type.csv")
df_omnibuses.columns
```

```python
set(df_omnibuses.type)
```

```python
types = ["Omnibuses", "Excursion buses", "Private buses"]
```

```python
df_omnibuses_filtered = df_omnibuses[df_omnibuses.type.isin(types)]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
for label, df in df_omnibuses_filtered.groupby("type"):
    df.plot("year", "number", ax=ax, label=label)
plt.xlim((2005, 2017))
plt.ylabel("number of vehicles")
plt.title("Number of vehicles over the years")
plt.legend()
plt.show()
```
